# Questions
### We will perform some data analysis on the conservation statuses of these species and investigate if there are any patterns or themes to the types of species that become endangered.

1. The two datasets are repertoring information about species in some geografical area (USA).
They both have 'scientific_name' with the same elements in them but some scientific names are showing multiple times, why?

2. There is 96.72% of missing values on species_info.conservation_status, why and what are we going to do with them?




### Answers
1. In the OBSERVATION dataset, the 'scientific_name' showing multiple times may due to different observations made by different person or different time or different location, ...

   In the SPECIES_INFO dataset, the 'scientific_name' can show multiple time also, this is due to difference in 'common_name' column, except for Canis Lupus where the 'conservation_status' are also different, therefore, since there is only one species (Canis Lupus) that have 2 different 'conservation_status', we are going to merge to 'scientific_names' in SPECIES_INFO to make it show only one time.
   For Canis Lupus, it has 3 occurence in our dataset, 2 of them states that conservation_status is 'Endangered', one states that it is 'In Recovery'.
   So we gonna keep the 'Endangered' status for Canis Lupus.

   Some species have also different 'consevation_status' but it is a NAN so we are going to keep the not NAN one.

2. For the nan values on species_info.conservation_status, the correction says that it means that those species are not under any conservation status

```python
# check species with more than one occurence
multiple_occured_species = species_info.scientific_name.value_counts()[species_info.scientific_name.value_counts() > 1].index
```

```python
# make a dataset with all the multiple occured species
recurent_scientific_name_in_species_info_dataset = pd.DataFrame()
for i in multiple_occured_species :
    recurent_scientific_name_in_species_info_dataset = pd.concat([recurent_scientific_name_in_species_info_dataset, species_info[species_info.scientific_name == i]])
```

```python
# how much of the multiple occurence dataset has non null conservation_status value
len(recurent_scientific_name_in_species_info_dataset) - recurent_scientific_name_in_species_info_dataset.conservation_status.isnull().sum()
```

```python
# here we can see that only Canis Lupus has 2 different conservation_status, others have either one status and/or NAN
recurent_scientific_name_in_species_info_dataset[recurent_scientific_name_in_species_info_dataset.conservation_status.isnull() == False]
```

```python
# species_info_merged_occurence
species_info2 = pd.DataFrame()
for i in species_info.scientific_name :
    temp = species_info[species_info.scientific_name == i]
    if len (temp) == 1 :
        species_info2 = pd.concat([species_info2, temp])
    elif len (temp) == 2 :
        if temp.conservation_status.isnull().sum() == 1 :
            species_info2 = pd.concat([species_info2, temp[temp.conservation_status.isnull() == False]])
        else : #temp.conservation_status.isnull().sum() == 2 or temp.conservation_status.isnull().sum() == 0
            species_info2 = pd.concat([species_info2, temp.iloc[:1]])
    else : #len (temp) == 3
        if temp.conservation_status.isnull().sum() == 1 or temp.conservation_status.isnull().sum() == 2 :
            species_info2 = pd.concat([species_info2, temp[temp.conservation_status.isnull() == False].iloc[:1]])
        else : #temp.iloc[0].scientific_name == 'Canis lupus' or temp.conservation_status.isnull().sum() == 3 or temp.conservation_status.isnull().sum() == 0
            species_info2 = pd.concat([species_info2, temp.iloc[:1]])
species_info2 = species_info2.drop_duplicates()

# let's drop the common_names column since it seems to be unecessary for our analysis
species_info2.drop(columns=['common_names'], inplace=True)
```

```python
# Now let's merge the 2 datasets observations and species_info2
data_merged = pd.merge(observations, species_info2, how = 'left', on='scientific_name')
data_merged.head()
```

```python
# function that count and show uniaue values from a dataset column
def show_unique_values (data, column) :
    print(f'Number of {column} {data[column].nunique()}')
    for i in data[column].unique() :
        print (f'- {i}')

# show barchart
def draw_countplot (data, column, color) :
    sns.countplot (data = data, x = column, color=color)
    plt.xticks(rotation=45)
    plt.show()
    plt.clf()
```

How many national parks do we have?

```python
show_unique_values(data_merged, 'park_name')
```

```python
How many categories do we have?
```

```python
show_unique_values(species_info, 'category')
draw_countplot (species_info, 'category', 'blue')
```

How many conservation status do we have

```python
show_unique_values(species_info, 'conservation_status')
draw_countplot (species_info, 'conservation_status', 'purple')
```

```python
import csv
# function that store the a column in the csv into a list
def column_to_list (csv_file_path, column_name) :
    output_list = []
    with open(csv_file_path) as insurance_csv :
        insurance_data = csv.DictReader(insurance_csv)
        for i in insurance_data :
            output_list.append(float(i[column_name]))
        return output_list
```

# Charges
Lets take an overall look on charges variable :
- show distribution
- median, IQR

```python
import numpy as np

charges = column_to_list("insurance.csv", "charges")
print (f"len = {len(charges)}, min = {min(charges)}, max = {max(charges)}")

q1, median, q3 = np.percentile(charges, [25, 50, 75])
print (f"q1 = {Q1}, median = {median}, q3 = {q3}")
```

```python
import matplotlib.pyplot as plt

plt.hist(charges, bins = 50)
plt.xlabel('Insurance cost')
plt.ylabel('Frequency')
plt.show()


plt.boxplot(charges)
plt.ylabel('Insurance cost')
plt.show()
```

#### What can we see ?
1. We can notice a right skewed distribution with 4 main groups that could be interesting for further analysis:
- group 1 around 0 and 15 000 dollars
- group 2 around 20 000 dollars
- group 3 around 40 000 dollars
- group 4 at 5000 dollars and more

2. We can notice that the group 4 and 5 are above the quartile q3
